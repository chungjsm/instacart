```python
# feature list: [aisle_id, department_id, days_since_prior_order, days_since_first_order, order_dow, order_hour_of_day]
schedule_dict = {}
for index, row in all_data_df.iterrows():
    if (row.user_id, row.product_id) not in schedule_dict:
        schedule_dict[(row.user_id, row.product_id)] = {"aisle_id":row.aisle_id, "department_id":row.department_id,
                                                        "days_since_prior_order":[], "day_number":[], "order_number":[],
                                                        "order_dow":[], "order_hour_of_day":[]}
    schedule_dict[(row.user_id, row.product_id)]["days_since_prior_order"].append(row.days_since_prior_order)
    schedule_dict[(row.user_id, row.product_id)]["day_number"].append(row.day_number)
    schedule_dict[(row.user_id, row.product_id)]["order_number"].append(row.order_number)
    schedule_dict[(row.user_id, row.product_id)]["order_dow"].append(row.order_dow)
    schedule_dict[(row.user_id, row.product_id)]["order_hour_of_day"].append(row.order_hour_of_day)
```

```python
features = []
for key in schedule_dict:
    user, product = key
    index = schedule_dict[key]
    length = len(index["order_dow"])
    for i in range(1, length):
        order = []
        order.append(index["aisle_id"])
        order.append(index["department_id"])
        order.append(index["days_since_prior_order"][i])
        order.append(index["day_number"][i] - index["day_number"][i-1])
        order.append(index["order_dow"][i])
        order.append(index["order_hour_of_day"][i])
        features.append(order)
```

```python
features = np.array(features)
enc = preprocessing.OneHotEncoder(categorical_features=[0,1])
enc.fit(features)
enc.n_values_
enc.feature_indices_
```

```python
feat_trans = np.concatenate((enc.transform(features[:,:2]).toarray(), features[:,2:]), axis=1)
```

```python
user20 = ord_df[(ord_df.eval_set=='test') & (ord_df.user_id==20)]
```

```python
model = sklearn.linear_model.LogisticRegression()
model.fit(feat_trans, np.array([np.random.choice([0,1]) for i in range(len(feat_trans))]))
```
